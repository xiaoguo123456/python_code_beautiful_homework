```python
with open('李白.txt',encoding='utf-8') as f:
    poems_text=f.read()
each_poem=poems_text.split('\n')

contend=[i.split(' ')[3] for i in each_poem if len(i.split(' '))==4]

all_str=''.join(contend)
lnum=[i for i in all_str if i.isalnum()]

print('李白诗词中总共有{}个汉字，其中不重复的汉字为{}个'.format(len(lnum),len(set(lnum))))

for i,i_str in enumerate(lnum):
      
      
      print(i_str,end='  ')
      if (i+1)%10==0:
        print()
      

    
    
```

```python
proty={i:lnum.count(i)/len(contend) for i in set(lnum)}
proty = {k:v for k,v in proty.items() if v > 0.1}
libai_proty=dict(sorted(proty.items(),key=lambda item:item[1],reverse=True ))
print('汉字-----频率')
for i in libai_proty.items():
    print(i[0],'    ',i[1])
```

```python
with open('杜甫.txt',encoding='utf-8') as f:
    dufu_poems_text=f.read()
dufu_poem=dufu_poems_text.split('\n')
dufu_contend=[i.split(' ')[3] for i in dufu_poem if len(i.split(' '))==4]
dufu_str=[i for i in (''.join(dufu_contend)) if i.isalnum()]
dufu_dict={i:dufu_str.count(i)/len(dufu_poem) for i in set(dufu_str)}
dufu_set=set([i[0] for i in dufu_dict.items() if i[1]>0.1])

both_like=dufu_set&set(libai_proty.keys())
print('两位诗人都喜欢用的汉字有{}个，它们为：'.format(len(both_like)))
for i,i_str in enumerate(both_like):
    print(i_str,end='  ')
    if (i+1)%10==0:
        print()
```

```python
libai_like=set(libai_proty.keys())-dufu_set
print('李白爱用但是杜甫不爱用的汉字有{}个，它们为：'.format(len(libai_like)))
for i,i_str in enumerate(libai_like):
    print(i_str,end='  ')
    if (i+1)%10==0:
        print()
```

```python
dufu_like=dufu_set-set(libai_proty.keys())
print('杜甫爱用但是李白不爱用的汉字有{}个，它们为：'.format(len(dufu_like)))
for i,i_str in enumerate(dufu_like):
    print(i_str,end='  ')
    if (i+1)%10==0:
        print()
```
